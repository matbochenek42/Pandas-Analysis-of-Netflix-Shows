```python
df = df.drop_duplicates() # remove duplicates
```

```python
df = df.drop(columns = "release_year") # column "release_year" deleted (release year is already in "date_added" column)
```

```python
# print(df.columns) # printing all columns

col = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'rating', 'duration', 'listed_in', 'description'] # all columns within list, except for "date_added" column

df[col] = df[col].astype(str) # changing columns data type to string type (except "date_added column")
```

```python
# deleting blank spaces from rows

df[col] = df[col].map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# replacing null values to "unknown"

col = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'rating', 'listed_in', 'description'] # all columns within list, except for "date_added" and "duration" column

df[col] = df[col].replace("nan", "Unknown")
```

```python
# formating a "date_added" column 

df["date_added"] = df["date_added"].str.strip() # deleting blank spaces in a "date_added" column

df["date_added"] = pd.to_datetime(df["date_added"], format="%B %d, %Y") # format change to yyyy-mm-dd

# format copied from chatgpt
```

```python
# creating two additional columns

df[["length", "unit"]] = df["duration"].str.split(' ', expand=True)

#df["length"].astype(int)

df["unit"] = df["unit"].str.strip()
```

```python
# setting index

df.set_index(["show_id"], inplace = True) 

df = df.sort_index()
```

```python
# saving cleaned data to a new file

df.to_csv(USER_PATH / "Pandas Analysis of Netflix Shows" / "data" / "data_cleaned.csv", index = True)
```

```python
# splitting some columns to get all the key data

df["cast"] = df["cast"].str.split(",")

df["country"] = df["country"].str.split(", ")

df["listed_in"] = df["listed_in"].str.split(", ")

```

```python
# assigning columns above to new dataframes

df_cast = df[["cast"]]

df_country = df[["country"]]

df_category = df[["listed_in"]]
```

```python
# unpivoting rows

df_cast = df_cast.explode("cast")

df_cast["cast"] = df_cast["cast"].str.lstrip() # lstrip was needed

df_country = df_country.explode("country")

df_category = df_category.explode("listed_in")
```

```python
# saving new dataframes to files

df_cast.to_csv(USER_PATH / "Pandas Analysis of Netflix Shows" / "data" / "cast.csv", index = True)

df_category.to_csv(USER_PATH / "Pandas Analysis of Netflix Shows" / "data" / "category.csv", index = True)

df_country.to_csv(USER_PATH / "Pandas Analysis of Netflix Shows" / "data" / "country.csv", index = True)
```
